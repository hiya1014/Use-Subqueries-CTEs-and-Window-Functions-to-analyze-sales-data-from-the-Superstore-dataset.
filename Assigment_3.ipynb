```python
conn = sqlite3.connect("superstore.db")
cursor = conn.cursor()
```

```python
df.to_sql(
    "superstore_raw",
    conn,
    if_exists="replace",
    index=False
)
```

```python
pd.read_sql(
    "SELECT * FROM superstore_raw LIMIT 5",
    conn
)
```

```python
cursor.execute("""
CREATE TABLE customers AS
SELECT DISTINCT
    [Customer ID] AS customer_id,
    [Customer Name] AS customer_name
FROM superstore_raw
""")

conn.commit()
```

```python
cursor.execute("""
CREATE TABLE orders AS
SELECT DISTINCT
    [Order ID] AS order_id,
    [Customer ID] AS customer_id,
    [Order Date] AS order_date,
    Sales
FROM superstore_raw
""")

conn.commit()
```

```python
cursor.execute("""
CREATE TABLE products AS
SELECT DISTINCT
    [Product ID] AS product_id,
    [Product Name] AS product_name,
    Category
FROM superstore_raw
""")

conn.commit()
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table'
""", conn)
```

```python
#Orders Above Average Sales (Subquery)
query = """
SELECT *
FROM orders
WHERE Sales >
(
    SELECT AVG(Sales)
    FROM orders
)
"""

pd.read_sql(query, conn)
```

```python
#Highest Order Per Customer
query = """
SELECT *
FROM orders o
WHERE Sales =
(
    SELECT MAX(o2.Sales)
    FROM orders o2
    WHERE o.customer_id = o2.customer_id
)
"""

pd.read_sql(query, conn)
```

```python
#Total Sales Per Customer (CTE)
query = """
WITH customer_sales AS
(
    SELECT
        customer_id,
        SUM(Sales) AS total_sales
    FROM orders
    GROUP BY customer_id
)

SELECT *
FROM customer_sales
"""

pd.read_sql(query, conn)
```

```python
#Customers Above Average Sales
query = """
WITH customer_sales AS
(
    SELECT
        customer_id,
        SUM(Sales) AS total_sales
    FROM orders
    GROUP BY customer_id
)

SELECT *
FROM customer_sales
WHERE total_sales >
(
    SELECT AVG(total_sales)
    FROM customer_sales
)
"""

pd.read_sql(query, conn)
```

```python
#Rank Customers by Sales
query = """
WITH customer_sales AS
(
    SELECT
        customer_id,
        SUM(Sales) AS total_sales
    FROM orders
    GROUP BY customer_id
)

SELECT
    customer_id,
    total_sales,
    RANK() OVER
    (
        ORDER BY total_sales DESC
    ) AS sales_rank
FROM customer_sales
"""

pd.read_sql(query, conn)
```

```python
#ROW_NUMBER()
query = """
SELECT
    customer_id,
    order_id,
    Sales,

ROW_NUMBER() OVER
(
PARTITION BY customer_id
ORDER BY Sales DESC
) AS row_num

FROM orders
"""

pd.read_sql(query, conn)
```

```python
#Top 5 Customers
query = """
WITH customer_sales AS
(
SELECT
customer_id,
SUM(Sales) AS total_sales
FROM orders
GROUP BY customer_id
)

SELECT *
FROM customer_sales
ORDER BY total_sales DESC
LIMIT 5
"""

pd.read_sql(query, conn)
```

```python
#Lowest 5 Customers
query = """
WITH customer_sales AS
(
SELECT
customer_id,
SUM(Sales) AS total_sales
FROM orders
GROUP BY customer_id
)

SELECT *
FROM customer_sales
ORDER BY total_sales
LIMIT 5
"""

pd.read_sql(query, conn)
```

```python
#Single Order Customers
query = """
SELECT
customer_id,
COUNT(*) AS order_count
FROM orders
GROUP BY customer_id
HAVING COUNT(*) = 1
"""

pd.read_sql(query, conn)
```

```python
#Highest Order Value Per Customer
query = """
SELECT
customer_id,
MAX(Sales) AS highest_order
FROM orders
GROUP BY customer_id
"""

pd.read_sql(query, conn)
```

```python
#JOIN + CTE + WINDOW FUNCTION
query = """
WITH customer_sales AS
(
SELECT
customer_id,
SUM(Sales) AS total_sales
FROM orders
GROUP BY customer_id
)

SELECT
c.customer_name,
cs.total_sales,

RANK() OVER
(
ORDER BY cs.total_sales DESC
) AS customer_rank

FROM customer_sales cs

JOIN customers c
ON cs.customer_id = c.customer_id

ORDER BY customer_rank
"""

pd.read_sql(query, conn)
```
